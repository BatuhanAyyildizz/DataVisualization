# Penguins

#### import operations

```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.info()
```

#### enguen

```python
sns.histplot(x ="flipper_length_mm",data =penguins ,binwidth =2 ,kde = True);
```

```python
sns.histplot(x ="flipper_length_mm",binwidth =2 ,hue ="species",data =penguins );
```

```python
sns.barplot(x="species",y="flipper_length_mm",data = penguins);
```

```python
sns.barplot(x="species",y="flipper_length_mm",hue ="sex",data = penguins);
```

```python
sns.boxplot(x ="species",y= "flipper_length_mm",hue = "sex",data=penguins);
```

```python
sns.violinplot(x ="species",y= "flipper_length_mm",hue = "sex",data=penguins);
```

```python
sns.scatterplot(x ="bill_length_mm",y= "bill_depth_mm",hue ="species",data =penguins );
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x ="bill_length_mm",y= "bill_depth_mm",hue ="species",col ="sex",data =penguins );
```

```python
sns.lmplot(x ="bill_length_mm",y= "bill_depth_mm",hue ="species",col ="sex",row = "island",data =penguins );
```

```python
sns.pairplot(penguins,hue ="species",height=3,markers = ["o","s","D"]);
```

```python
sns.pairplot(penguins,hue ="species",height=3,markers = ["o","s","D"] ,diag_kind= "hist");
```
